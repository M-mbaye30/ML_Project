            ***premier dataset***

```python
#on importe les librairies pandas et matplotlib
import pandas as pd
import matplotlib as plt
```

```python
#on charge le fichier csv
df = pd.read_csv("multilingual_reviews.csv")
# on le visualise pour s'assurer qu'il est bien chargé
df
```

on voit que le dataset contient 40000 lignes et 8 colonnes


```python
# on filtre les données pour ne garder que les avis en anglais
df_en = df[df['language'] == 'en']
#on visualise le résultat aussi
df_en
```

```python
# on selectionne que les colonne review_body et product_category
df_selected = df_en[["review_body", "product_category"]]
#on visualise le résultat
df_selected
```

```python
#on renomme les colonnes 
df_selected= df_selected.rename(columns= {"review_body": "texts", "product_category": "classes"})
df_selected
```

```python
#on sauvegarde le fichier en csv sous le nom reviews_en.csv
df_selected.to_csv("data/reviews_en.csv", index=False)
df_selected
```

après avoir filtré le dataset, on peut voir que le nombre ligne a diminué de 40000 à 10000 de meme que les colonnes de 8 à 2

              ***deuxième dataset***

```python
#on charge le fichier csv 
df = pd.read_csv("multilingual_reviews.csv")
df
```

```python
#on filtre les colonnes pour ne garder que review_body et language
df_filtred = df[["review_body", "language"]]
#on visualise le résultat
df_filtred
```

```python
#on renomme les colonnes qu'on a décidé de garder en texts et lg
df_filtred= df_filtred.rename(columns={"review_body": "texts", "language": "lg"})
df_filtred
```

```python
#on sauvegarde le fichier en csv sous le nom multilingual_dataset.csv dans le dossier data index=False pour ne pas garder l'index 
df_filtred.to_csv("data/multilingual_dataset.csv", index=False)
```

pour notre second dataset le nombre de ligne reste le meme mais après l'avoir filtré, par concéquent on a pas le meme nombre de colonnes car on a gardé que deux colonnes.
